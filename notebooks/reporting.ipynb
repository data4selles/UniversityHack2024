# Reporting

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from joblib import load
```

```python
metrics = pd.read_csv('../06_model_output/metrics/final_metrics.txt', header=None, index_col=0, names=['Run_Time', 'RMSE'])
metrics
```

```python
train = pd.read_excel('../05_model_input/train.xlsx') 
y = train.PRODUCTO_1
X = train.drop('PRODUCTO_1', axis=1)
```

```python
prediction_rf = rf.predict(X)
error_rf = prediction_rf - y

prediction_dt = dt.predict(X)
error_dt = prediction_dt - y

prediction_gb = gb.predict(X)
error_gb = prediction_gb - y

prediction_xgb = xgb.predict(X)
error_xgb = prediction_xgb - y
```

Leemos los modelos entrenados y vemos las predicciones que dan, a pesar de que a simple vista se ve que el que mejor ajusta el conjunto de entrenamiento es XGBoost

```python
rf = load('../06_model_output/models/random_forest.joblib')
dt = load('../06_model_output/models/decision_tree.joblib')
gb = load('../06_model_output/models/gradient_boost.joblib')
xgb = load('../06_model_output/models/xgboost.joblib')
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].scatter(y, prediction_rf)
axs[0, 0].plot([y.min(), y.max()], [y.min(), y.max()], '--r', lw=2)
axs[0, 0].set_xlabel('Valores Reales')
axs[0, 0].set_ylabel('Predicciones')
axs[0, 0].set_title('Predicciones vs valores reales Random Forest')

axs[0, 1].scatter(y, prediction_dt)
axs[0, 1].plot([y.min(), y.max()], [y.min(), y.max()], '--r', lw=2)
axs[0, 1].set_xlabel('Valores Reales')
axs[0, 1].set_ylabel('Predicciones')
axs[0, 1].set_title('Predicciones vs valores reales Decision Tree')

axs[1, 0].scatter(y, prediction_gb)
axs[1, 0].plot([y.min(), y.max()], [y.min(), y.max()], '--r', lw=2)
axs[1, 0].set_xlabel('Valores Reales')
axs[1, 0].set_ylabel('Predicciones')
axs[1, 0].set_title('Predicciones vs valores reales Gradient Boosting')

axs[1, 1].scatter(y, prediction_xgb)
axs[1, 1].plot([y.min(), y.max()], [y.min(), y.max()], '--r', lw=2)
axs[1, 1].set_xlabel('Valores Reales')
axs[1, 1].set_ylabel('Predicciones')
axs[1, 1].set_title('Predicciones vs valores reales XGBoost')

plt.tight_layout()
plt.show()
```
